## Exploratory Data Analysis of Automobile.txt

```python
# Import all libraries needed
import pandas as pd
import numpy as np
import missingno
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the file as a dataframe csv
df = pd.read_csv('automobile.txt')
df
```

```python
# We notice already there are values with '?'
# We want to replace these values with NaN
df_data = df.replace('?', np.NAN)

# We now count how many values in the dataframe were replaced
df_data.isnull().sum()
```

```python
df_data
```

```python
# Create a visualization of the missing data
missingno.matrix(df_data, figsize = (15,5))
plt.show()
```

```python
# Find the % count of missing data
missing_values_count = df_data.isnull().sum()
missing_values_count
missing_values_count / len(df_data) * 100
```

### Approach to filling of Missing data
I want to determine the datas current descriptive state.
I believe that with all columns, except for the 'num_of_doors'; I can fill with a normalised mean.
After having filled the data; I will then reperform a descriptive state of the data.
If any values vary by an extreme amount, I will reassess my approach to filling the missing data

```python
# Current state of the data with its missing values
df_data.describe()
```

#### Filling in the Missing Data

```python
# Create a temporary dataframe where we populate the missing values
# fill missing data of normalised losses
df_temp = df[df['normalized-losses']!='?']
normalised_mean = df_temp['normalized-losses'].astype(int).mean()
df['normalized-losses'] = df['normalized-losses'].replace('?',normalised_mean).astype(int)

# fill missing data of price
df_temp = df[df['price']!='?']
normalised_mean = df_temp['price'].astype(int).mean()
df['price'] = df['price'].replace('?',normalised_mean).astype(int)

# fill missing data of horsepower
df_temp = df[df['horsepower']!='?']
normalised_mean = df_temp['horsepower'].astype(int).mean()
df['horsepower'] = df['horsepower'].replace('?',normalised_mean).astype(int)

# fill missing data of peak rpm
df_temp = df[df['peak-rpm']!='?']
normalised_mean = df_temp['peak-rpm'].astype(int).mean()
df['peak-rpm'] = df['peak-rpm'].replace('?',normalised_mean).astype(int)

# fill missing data of bore
df_temp = df[df['bore']!='?']
normalised_mean = df_temp['bore'].astype(float).mean()
df['bore'] = df['bore'].replace('?',normalised_mean).astype(float)

# fill missing data of stroke
df_temp = df[df['stroke']!='?']
normalised_mean = df_temp['stroke'].astype(float).mean()
df['stroke'] = df['stroke'].replace('?',normalised_mean).astype(float)

# Fill missing data category Number of doors with the mode of the column i.e. Four
df['num-of-doors'] = df['num-of-doors'].replace('?','four')

df
```

```python
# I now perform the same descriptive state on the temp dataframe
df.describe()
```

```python
# Checking to see if any values need to be converted
df.dtypes
```

#### Changing Data Formats
We have 2 columns that include mpg (Miles per Gallon). This data bares no meaning to us, as we use km per litre. We need to convert the data in a meaningful manner for users.

```python
df[['city-mpg']] = 235/df[['city-mpg']]
df.rename(columns={'city-mpg':'city_L/100km'}, inplace=True)
df
```

```python
df[['highway-mpg']] = 235/df[['highway-mpg']]
df.rename(columns={'highway-mpg':'highway_L/100km'}, inplace=True)
df
```

#### Heatmaps

```python
corr = df.corr()
plt.figure(figsize=(20,9))
a = sns.heatmap(corr, annot=True, fmt='.2f')
plt.title('Heatmap showing how each category relates to one another')
plt.show()
```

```python
df_group = df[['drive-wheels', 'body-style', 'price']]
# Find average price across 3 drive-wheel types
group_av = df_group.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
# Store all entries into a table
table = group_av.pivot(index='drive-wheels', columns='body-style')
# Fill missing values with 0
table = table.fillna(0)
table
# We now use the table we created to show a heatmap
sns.heatmap(table)
plt.title('Heatmap showing drive-wheels and body-style to price')
plt.xlabel('Body Style')
plt.show()
```
